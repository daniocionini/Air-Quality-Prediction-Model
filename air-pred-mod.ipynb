```python
import pandas as pd
```

```python
df = pd.read_csv('./air_quality.csv', sep=';')
df
```

```python
df.shape
```

```python
# identify where the missing values are in the dataset
df.isna().any()
```

```python
# find % of missing values per single attribute in the dataset
nan = df.isna().sum()
tot = df.count()
perc = (nan*100)/tot
perc
```

```python
data = df.dropna(subset=['Date'])
data
```

```python
data = data.drop(['Unnamed: 15', 'Unnamed: 16'], axis=1)
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
data['CO(GT)'] = (data['CO(GT)'].replace(',','.',regex=True).astype(float))
data['C6H6(GT)'] = (data['C6H6(GT)'].replace(',','.',regex=True).astype(float))
data['T'] = (data['T'].replace(',','.',regex=True).astype(float))
data['RH'] = (data['RH'].replace(',','.',regex=True).astype(float))
data['AH'] = (data['AH'].replace(',','.',regex=True).astype(float))
data['Date'] = pd.to_datetime(data['Date'], format='%d/%m/%Y')
data = data.loc[data["CO(GT)"] > -10] #remove all rows where CO is registered as -200 (missing value)
data.info()
```

```python
data.duplicated().sum()
```

```python
data.head()
```

```python
f = data.describe()
f
```

```python
import matplotlib.pyplot as plt 
from scipy.stats import norm
import numpy as np
import seaborn as sns
```

```python
s = f.drop(['count','min'])
s.plot(kind='bar',figsize=(10,8))
```

```python
x = np.arange(-10, 10, 0.01)
mean = f['CO(GT)'].iloc[[1]]
std = f['CO(GT)'].iloc[[2]]
median = f['CO(GT)'].iloc[[5]]
```

```python
plt.figure(figsize=(10,8))
plt.plot(x, norm.pdf(x, mean, std))
plt.axvline(float(mean), 0, 0.95, label='Mean', c='red')
plt.axvline(float(median), 0, 0.93, label='Median', c='green')
plt.legend()
plt.show()
```

The mean > median meaning the distribution of **CO** is positively skewed
